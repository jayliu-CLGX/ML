```python
import pandas as pd
import numpy as np
```

```python
sv_cbsa19q1 = pd.read_csv('gs://results_idap_analytics_us-saarisk-uat/hpi_locationinc/cbsa_scout_vision_q1_2019.csv')
sv_cbsa19q1.shape
```

```python
sv_cbsa22q2 = pd.read_csv('gs://results_idap_analytics_us-saarisk-uat/hpi_locationinc/cbsa_scout_vision.csv')
sv_cbsa22q2.shape
```

```python
sv_cbsa22q2
```

```python
sv_cbsa22q2.drop_duplicates(subset=['cbsa5'])
```

```python
sv_cbsa19q1['hpa_1yr_pred']=pd.to_numeric(sv_cbsa19q1['mhv_pred_1y'], errors='coerce').astype('Int64')/sv_cbsa19q1['mhv'].astype('Int64')-1
```

```python
sv_cbsa19q1_hpa=sv_cbsa19q1[['cbsa5', 'cbsa_title','hpa_1yr_pred','hpa_3yr_pred']]
```

```python
sv_cbsa22q2['hpa_1yr_hist']=pd.to_numeric(sv_cbsa22q2['mhv_obs_2019_4'], errors='coerce').astype('Int64')/sv_cbsa22q2['mhv_obs_2018_4'].astype('Int64')-1
sv_cbsa22q2['hpa_3yr_hist']=pd.to_numeric(sv_cbsa22q2['mhv_obs_2021_4'], errors='coerce').astype('Int64')/sv_cbsa22q2['mhv_obs_2018_4'].astype('Int64')-1
```

```python
sv_cbsa22q2
```

```python
sv_cbsa22q2_hpa_his=sv_cbsa22q2[['cbsa5', 'cbsa_title','hpa_1yr_hist','hpa_3yr_hist']]
```

```python
sv_cbsa22q2_hpa_his
```

```python
sv_cbsa19q1_hpa.set_index('cbsa5', inplace=True, drop=True)
```

```python
sv_cbsa22q2_hpa_his.set_index('cbsa5', inplace=True, drop=True)
```

```python
sv_cbsa22q2_hpa_his
```

```python
sv_cbsa19q1_hpa_new=pd.merge(sv_cbsa19q1_hpa, sv_cbsa22q2_hpa_his, how="left", left_index=True, right_index=True)
```

```python
sv_cbsa19q1_hpa_new['error_1yr']=sv_cbsa19q1_hpa_new['hpa_1yr_pred']/sv_cbsa19q1_hpa_new['hpa_1yr_hist']-1
sv_cbsa19q1_hpa_new['error_abs_1yr']=sv_cbsa19q1_hpa_new['error_1yr'].abs()

sv_cbsa19q1_hpa_new['error_3yr']=sv_cbsa19q1_hpa_new['hpa_3yr_pred']/sv_cbsa19q1_hpa_new['hpa_3yr_hist']-1
sv_cbsa19q1_hpa_new['error_abs_3yr']=sv_cbsa19q1_hpa_new['error_3yr'].abs()
```

```python
sv_cbsa19q1_hpa_new
```

```python
sv_cbsa19q1_hpa_new.describe()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
```

```python
sv_cbsa19q1_hpa_new.hist(column='error_3yr')
```

```python
sv_cbsa19q1_hpa_new.sort_values(by='error_abs_1yr', ascending=False)
```
